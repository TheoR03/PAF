```
# nvdia-smi

# Importing relevant libraries

import numpy as np
np.random.seed(123)
import cv2 as cv

from keras import backend as K 
K.set_image_dim_ordering('th') 

from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten
from keras.layers import Convolution2D, MaxPooling2D
from keras.utils  import np_utils

import glob

import pandas as pd
import scipy.misc
import cv2
import io
import zipfile
import imageio
from google.colab import drive
from google.colab import files


drive.mount('/content/drive/')


uploaded = files.upload() #csv files


# Creating training and testing sets from database


file_list = []

path = glob.glob('/content/drive/My Drive/Colab Notebooks/data2/*.jpg')

for img in path:
  file_list.append(cv.imread(img))

  
(y_size,x_size,n_channels) = (1, 2, 3) # just initialising
#file_list = uploaded.values() #list of pictures


x_train = np.zeros((64,n_channels,y_size,x_size)) # training set initialisation
for i in range (700):
  print(file_list[i].shape
  param = file_list[i].shape
  (y_size,x_size,n_channels)=(param[0],param[2],param[3])
  img_in = file_list[i].astype(float)/255.0
  x_train[i,:,:,:] = np.transpose(img_in,(0,n_channels,y_size,x_size))

  
y_train = np.array(pd.read_csv(io.StringIO(uploaded['y_train.csv'].decode('utf-8')))) # training target values


x_test = np.zeros((64,n_channels,y_size,x_size)) # testing set initialisation
for i in range (700, len(file_list)):
  param = file_list[i].shape
  (y_size,x_size,n_channels)=(param[0],param[2],param[3])
  img_in = file_list[i].astype(float)/255.0
  x_test[i,:,:,:] = np.transpose(img_in,(0,n_channels,y_size,x_size))
  
  
  
y_test = np.array(pd.read_csv(io.StringIO(uploaded['y_test.csv'].decode('utf-8')))) #testing target values


# 1. Building the model architecture

model = Sequential()

model.add(Convolution2D(64, (3, 3), activation='relu', input_shape=(1,28,28)))
# first hidden convolutional neutrons layer, 64 neutrons 
model.add(Convolution2D(32, (3, 3), activation='relu'))
# second hidden convolutional neutrons layer, 32 neutrons whereas 64 in previous layer
model.add(Flatten())
# flatten to get proper dimension
model.add(Dense(1, activation='sigmoid'))
# last layer


# 2. Compiling the model
model.compile(loss='binary_crossentropy', optimizer='rmsprop', metrics=['accuracy'])

# 3. Training the model
model.fit(x_train, y_train, epochs=20, batch_size=64)

# 4. Testing the model
y_pred = model.predict_classes(x_test, batch_size=64)

# 5. Number of positions that differ between prediction and ground truth
print(sum(abs(y_pred-y_test)))

# 6. Assessing the model
score = model.evaluate(x_test, y_test, batch_size=64)

```
